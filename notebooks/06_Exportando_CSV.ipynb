## 1. Exportando tabelas da camada gold para csv

###1.1 Script de importação para streamlit(csv)

```python
# Lista das tabelas essenciais para o app
tabelas = ["fct_vendas", "dim_produtos", "dim_clientes", "dim_calendario"]
```

```python
# Pasta temporária no dbfs
path_export = "dbfs:/FileStore/export_streamlit/"
dbutils.fs.mkdirs(path_export)

print("Gerando CSVs para Streamlit...")
```

```python
for tabela in tabelas:
    # Read the gold table
    df = spark.read.table(f"workspace_ecommerce.gold.{tabela}")

    # Save as a single CSV file with header
    df.coalesce(1).write.format("csv").option("header", "true").mode("overwrite").save(f"{path_export}{tabela}")

    # Find the generated CSV file (part-000...)
    arquivos = dbutils.fs.ls(f"{path_export}{tabela}/")
    csv_file = [f.name for f in arquivos if f.name.endswith(".csv")][0]

    # Generate download link
    workspace_url = spark.conf.get("spark.databricks.workspaceUrl")
    print(
        f"{tabela.upper()}: https://{workspace_url}/files/export_streamlit/{tabela}/{csv_file}?o={workspace_url.split('.')[0].split('-')[1]}"
    )
```
